```python
all_r = pd.DataFrame(columns = uniquekeywords,index = range(1990,2017))
```

```python
for i, word in enumerate(uniquekeywords):
    all_r[word] = top_keywords[top_keywords.keyword == word].groupby(by = 'year').count()
    #print all_r[word]
    if np.mod(i,1000) == 0:
        print 'word',i,'complete'
```

```python
all_r = all_r.fillna(0)
#number of occurances in each year
d = top_keywords.groupby('year').keyword.count().to_frame().keyword.tolist()
all_r.shape
```

```python
all_bursts = pd.DataFrame(columns=['begin','end','weight'])
#define variables
s = 2         #resolution of state jumps; higher s --> fewer but stronger bursts
gam = 0.1     #difficulty of moving up a state; larger gamma --> harder to move up states, less bursty
n = len(d)    #number of timepoints

#loop through unique words
for i, word in enumerate(uniquekeywords):

    r = list(all_r.loc[:,word].astype(int))

    #find the optimal state sequence (using the Viterbi algorithm)
    [q,d,r,p] = bd.burst_detection(r,d,n,s,gam,smooth_win=0)

    #enumerate the bursts
    bursts = bd.enumerate_bursts(q, word)

    #find weight of each burst
    bursts = bd.burst_weights(bursts, r, d, p)

    #add the bursts to a list of all bursts
    all_bursts = all_bursts.append(bursts, ignore_index=True)
    
    #print a progress report every 100 words
    if np.mod(i,1000)==0:
        print('word',i,'complete')

all_bursts.sort_values(by='weight', ascending=False).head(10)
```

```python
all_bursts.sort_values(by='weight', ascending=False).head(20)
```

```python
tmp = all_bursts.sort_values(by='weight', ascending=False).head(10)[['label','weight']]
tmp['rank'] = tmp['weight'].div(np.sum(tmp['weight']))
label_rank = {}
label_weight_index = {}
for index, row in tmp.iterrows():
    label_weight_index[row['label']] = row['weight']
    label_rank[row['label']] = row['rank']
print label_weight_index
print label_rank
```

```python
#using top keywords data from above
uniquekeywords = list(top_keywords.keyword.unique())
M_keywords = np.zeros(shape = (len(uniquekeywords),len(uniquekeywords)))
```

```python
def keywordcooccur(x):
    words = list(x)
    if len(words)<2:
        return 1
    else:
        tmpcombs = list(combinations(words, 2))
        for tmpcomb in tmpcombs:
            first_word = tmpcomb[0]
            second_word = tmpcomb[1]
            first_index = uniquekeywords.index(first_word)
            second_index = uniquekeywords.index(second_word)
            M_keywords[first_index,second_index]+=1
            M_keywords[second_index,first_index]+=1
        return 2
```

```python
top_keywords.groupby('UID').keyword.apply(keywordcooccur)
print "done"
```

```python
M_keywords.sum()
```

```python
Z = linkage(M_keywords, 'ward')
```

```python
plt.figure(figsize=(80,10))
plt.title('Hierarchical Clustering Dendrogram', fontsize=20)
plt.ylabel('distance', fontsize=16)

# call dendrogram to get the returned dictionary 
# (plotting parameters can be ignored at this point)
R = dendrogram(
                Z,
                no_plot=True,
                )

# create a label dictionary
temp = {R["leaves"][ii]: uniquekeywords[ii] for ii in range(len(R["leaves"]))}
def llf(xx):
    return temp[xx]



dendrogram(
            Z,
            leaf_label_func=llf,
            leaf_rotation=90.,  # rotates the x axis labels
            leaf_font_size=8.            )
plt.show()
```

```python
G = nx.from_numpy_matrix(M_keywords)
labelsdict = {}
for i in range(len(uniquekeywords)):
    labelsdict[i]=uniquekeywords[i]

G = nx.relabel_nodes(G,labelsdict,copy=False)
```

```python
#get top values in a centrality measure
def get_top_values(dictionary, top):
    items = list(dictionary.items())
    items = sorted(items,reverse=True, key=lambda x: x[1])
    return items[:top]
```

```python
# deg cen cities scaled
deg_cen_keywords = nx.degree_centrality(G)
get_top_values(deg_cen_keywords,10)
```

## MDS plot

```python
from sklearn import manifold
from sklearn.metrics import euclidean_distances
```

```python
similarities = euclidean_distances(M_keywords)
```

```python
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=123,
                   dissimilarity="precomputed", n_jobs=1)
pos = mds.fit(similarities).embedding_
pos *= np.sqrt((M_keywords ** 2).sum()) / np.sqrt((pos ** 2).sum())
```

```python
fig = plt.figure(figsize = (15,15))
plt.subplots_adjust(bottom = 0.1)
plt.scatter(
    pos[:, 0], pos[:, 1]
    )
for label, x, y in zip(uniquekeywords, pos[:, 0], pos[:, 1]):
    plt.annotate(
        label,
        xy = (x, y), xytext = (-20, 20),
        textcoords = 'offset points', ha = 'right', va = 'bottom',
        bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))

plt.show()
```

## Kaplan - Meier curve estimation

```python
def getstart(x):
    x = list(x)
    return np.min(x)
def getend(x):
    x = list(x)
    return np.max(x)
```

```python
tmpstart = keyword_data.groupby('keyword').year.apply(getstart).to_frame()
tmpstart['keyword'] = tmpstart.index
tmpstart.head()
```

```python
newkeywords = tmpstart.groupby('year').keyword.nunique().to_frame()
newkeywords['year'] = newkeywords.index
newkeywords.columns = ['numnew','year']
newkeywords.head()
```

```python
tmpend = keyword_data.groupby('keyword').year.apply(getend).to_frame()
tmpend['keyword'] = tmpend.index
tmpend.head()
```

```python
kaplan_data = tmpstart.merge(tmpend, how = 'left',on = 'keyword')
kaplan_data['istimeline'] = (kaplan_data['year_x']>2002)&(kaplan_data['year_x']<2013)
kaplan_data['istimeline'] = kaplan_data['istimeline'].astype(int)
kaplan_data['survived'] = kaplan_data['year_y']<2013
kaplan_data['duration'] = kaplan_data['year_y'] - kaplan_data['year_x'] + 1
kaplan_data['survived'] = kaplan_data['survived'].astype(int)
kaplan_data.head()
```

```python
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
```

```python
tmpk = kaplan_data[kaplan_data['istimeline'] == 1]
T = tmpk["duration"]
E = tmpk["survived"]

kmf.fit(T, event_observed=E)
```
